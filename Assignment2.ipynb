##### 5. Write lowest and highest rank record to new file

##### 'highest_len_satus_clm': [{ row_num : len_value} , { row_num : len_value}, { row_num : len_value}]
##### 'highest_rank_rec_val': ?,
##### 'highest_len_row': { row_nm : len_of _row }
##### }

```python
import openpyxl
#1 

def get_columns(file_name):
    wb = openpyxl.load_workbook(file_name)
    sheet = wb.active
    cols = sheet.max_column
    l = []
    for i in range(1,cols+1):
        cell = sheet.cell(row=1,column = i)
        l.append(cell.value)
    return l    

 

#2

def get_column_length(file_name):
    wb = openpyxl.load_workbook(file_name)
    sheet = wb.active
    cols = sheet.max_column
    d = {}
    for i in range(1,cols+1):
        cell = sheet.cell(row=1,column = i)
        d[cell.value]=len(cell.value)
    return d    



#3

def highest_length(file_name,col_name):
    wb = openpyxl.load_workbook(file_name)
    sheet = wb.active
    cols = sheet.max_column
    rows = sheet.max_row
    l = []
    for i in range(1,cols+1):
        cell = sheet.cell(row=1,column = i)
        if cell.value == col_name:
            for j in range(1,rows+1):
                c = sheet.cell(column = i, row=j)
                l.append(c.value)
               
        
    m = max(l)
    d={}
    for i in range(1,cols+1):
        cell = sheet.cell(row=1,column = i)
        if cell.value == col_name:
            for j in range(1,rows+1):
                c = sheet.cell(column = i, row=j)
                if c.value == m:
                    d[sheet.cell(column = 1, row = j).value]=c.value  # {rank:value of corresponding status col}
                
    return d
    
    


#4

def highest_rank_record(file_name):
    wb = openpyxl.load_workbook(file_name)
    sheet = wb.active
    rows = sheet.max_row
    cols = sheet.max_column
    l=[]
    for i in range(1,cols+1):
        cell = sheet.cell(row=101,column = i)
        l.append(cell.value) 
    return l
   

#5

def lowest_and_highest_rank(filename,new_file):
    wb = openpyxl.load_workbook(filename)
    sheet = wb.active
    rows = sheet.max_row
    cols = sheet.max_column
    low = []
    high = []
    fields = []
    for i in range(1,cols+1):
        cell = sheet.cell(row=101,column = i)
        high.append(cell.value) 
    for i in range(1,cols+1):
        cell = sheet.cell(row=2,column = i)
        low.append(cell.value)
    for i in range(1,cols+1):
        cell = sheet.cell(row=1,column = i)
        fields.append(cell.value)
        
    wb1 = openpyxl.load_workbook(new_file)
    sheet1 = wb1.active
    print(cols," ",rows)
    print(len(high))
    for i in range(1,cols+1):
        cell1 = sheet1.cell(row=1,column = i)
        cell1.value=fields[i-1]

    for i in range(1,cols+1):
        cell1 = sheet1.cell(row=2,column = i)
        cell1.value=low[i-1]   
      
    for i in range(1,cols+1):
        cell1 = sheet1.cell(row=3,column = i)
        cell1.value=high[i-1]
    wb1.save(new_file)
lowest_and_highest_rank('covid_data.xlsx','new_file.xlsx')    
```

```python
try:
    print("Get Columns: ")
    res = get_columns('covid_data.xlsx')
    print(res)

except Exception:
    print("Something went wrong in get_columns")
    
try:
    print("Get column lengths: ")
    res = get_column_length('covid_data.xlsx')
    print(res)
except Exception:
    print("Something went wrong in get_column_length")  
    
try:
    print("Highest Length cell in Status column: ")
    res = highest_length('covid_data.xlsx','Status')
    print(res)
except Exception:
    print("Something went wrong in highest_length")  
    
try:
    print("Highest Rank Record: ")
    res = highest_rank_record('covid_data.xlsx')
    print(res)
except Exception:
    print("Something went wrong in highest_rank_record")  

try:
    print("Lowest and Highest rank record in new file: ")
    lowest_and_highest_rank('covid_data.xlsx','new_file.xlsx')
    print("Successfully added records in new file!")
except Exception:
    print("Something went wrong in lowest_and_highest_rank")     
```
